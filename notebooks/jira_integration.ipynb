## Imports

```python
from jira import JIRA
```

## Personal token

```python
server_url = "" # PLACE THE SERVER URL HERE
token = "" # PLACE YOUR PERSONAL TOKEN HERE
```

## Connection

```python
jira = JIRA(server=server_url, token_auth=token, )
print(f"Connected to JIRA using: {jira}")
```

## Searching JIRA issues

```python
issues = jira.search_issues(
  """project = "Project X"
        AND (status = "In Progress" OR status = "In Review" OR status = "QA review" )
        AND sprint IN openSprints()
        ORDER BY created DESC"""
)

for issue in issues:
    print(f"{issue.key} - {issue.fields.summary}")
```

## Updating issues

```python
issue = jira.issue("PROJ-5500")
issue.update(description="New Description")
print("Description:", issue.fields.description)
```

```python
print(issue.fields.status.name)
transitions = jira.transitions(issue)
for t in transitions:
    print(t)

jira.transition_issue(issue, "To review")
issue = jira.issue("PROJ-5500")
print(issue.fields.status.name)
```
