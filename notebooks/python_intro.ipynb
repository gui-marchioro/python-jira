## Variables

Variables store information.


```python
issue_key = "PROJ-123"
status = "Open"

print(issue_key)
print(status)
```

## Lists

Lists allow us to store multiple values.


```python
# List creation
issues = ["PROJ-1", "PROJ-2", "PROJ-3"]
print(issues)

```

```python
# Adding items to a list
issues += ["PROJ-4"]
issues.append("PROJ-5")
print(issues)
```

```python
# Getting items from a list
i = issues[0]
print(i)
i = issues[-1]
print(i)
```

```python
# Remove items from a list
issues.remove("PROJ-2")
print(issues)
i = issues.pop()
print(issues)
```

```python
# Update item from a list
issues[0] = "PROJ-123"
print(issues)
```

## Dictionaries

Dictionaries store information as **key â†’ value** pairs.
This is how Jira data usually looks in Python.


```python
# Dictionary creation
issue = {
    "key": "PROJ-1",
    "status": "Open",
    "priority": "High"
}
print(issue)
```

```python
# Accessing a value
print(issue["key"])
```

```python
# Adding a new key-value pair
issue["assignee"] = "Alex"
print(issue)
```

```python
# Updating an existing value
issue["status"] = "In Progress"
print(issue)
```

```python
# Deleting a key-value pair
removed_value = issue.pop("assignee")
print(issue)
print("Removed value:", removed_value)
```

## If statement

Python can make decisions based on conditions.


```python
if issue["status"] == "Open":
    print("This issue needs attention")
elif issue["status"] == "In Progress":
    print("This issue is in progress")
```

## Loops

Loops allow Python to repeat actions for multiple items.


```python
issues = [
    {"key": "PROJ-1", "status": "Open"},
    {"key": "PROJ-2", "status": "Done"},
    {"key": "PROJ-3", "status": "Open"}
]

for issue in issues:
    if issue["status"] == "Open":
        print(issue["key"])
```
